# Data Exploration for Face Mask Detection

This notebook demonstrates how to load and explore the dataset used for Face Mask detection.

```python
import os
import matplotlib.pyplot as plt
import cv2
import random
```

```python
DATA_DIR = '../dataset'
classes = ['with_mask', 'without_mask']

# Check class distribution
for cls in classes:
    if os.path.exists(os.path.join(DATA_DIR, cls)):
        print(f"{cls} count:", len(os.listdir(os.path.join(DATA_DIR, cls))))
    else:
        print(f"Directory {cls} not found.")
```

```python
def show_random_images(classes, num_images=3):
    plt.figure(figsize=(10, 5))
    for i, cls in enumerate(classes):
        cls_dir = os.path.join(DATA_DIR, cls)
        if not os.path.exists(cls_dir):
            continue
        
        images = os.listdir(cls_dir)
        sample_img = random.choice(images)
        img_path = os.path.join(cls_dir, sample_img)
        
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        plt.subplot(1, len(classes), i + 1)
        plt.imshow(img)
        plt.title(f"{cls}")
        plt.axis('off')
    plt.show()

show_random_images(classes)
```
